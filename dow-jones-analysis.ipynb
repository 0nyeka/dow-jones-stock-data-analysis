```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('dow_jones_index.csv', parse_dates=['date'])
df = df.iloc[:, [1,2,3,4,5,6,7]]
df.set_index('date', inplace=True)
df
```

```python
ibm=df[df['stock']=='IBM']
ibm
```

```python
ibm['open'] = ibm['open'].str.replace('$', '', regex=False)
ibm['open'] = ibm['open'].astype(float)
ibm['open'].mean()
```

```python
plt.plot(ibm['open'])
plt.title("IBM Opening Stock Prices")
plt.xlabel("Date")       # Label for the horizontal axis
plt.ylabel("Price ($)")   # Label for the vertical axis
plt.show()
```

```python
wmt=df[df['stock']=='WMT']
wmt
```

```python
wmt['close'] = wmt['close'].str.replace('$', '', regex=False)
wmt['close'] = wmt['close'].astype(float)
wmt['close'].mean()
```

```python
plt.plot(wmt['close'])
plt.title("Walmart Closing Stock Prices")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.show()
```

```python
df_sum = df.groupby(['stock']).sum()
df_sum.plot(kind='pie', y='volume')
plt.title("Total Trading Volume by Stock")
plt.ylabel("")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
df_date = df.sort_index()
ibm_bac = df_date.loc[(df_date['stock']=='IBM') | (df_date['stock']=='BAC')]
ibm_bac_sum = ibm_bac.groupby(['stock']).sum()
ibm_bac_sum
```

```python
ibm_bac_sum.plot(kind='pie', y='volume', autopct='%1.1f%%')

plt.title("Trading Volume Comparison: IBM vs BAC") # Adds the title
plt.ylabel("") # Removes the 'volume' label from the left side
plt.show()
```

```python
wmt_bac_csco = df_date.loc[(df_date['stock']=='WMT') | (df_date['stock']=='BAC') | (df_date['stock']=='CSCO')]
wmt_bac_csco_sum = wmt_bac_csco_sum = wmt_bac_csco.groupby(['stock']).sum()
print(wmt_bac_csco_sum)
wmt_bac_csco_sum.plot(kind='pie', y='volume')
plt.show()
```
